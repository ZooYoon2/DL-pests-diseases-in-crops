```python
import os,cv2,keras
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator

TomatoModel = keras.models.load_model("Model_5class_test2")
category = {0:'leafMold',1:'leafSpot',2:'normal',3:'trifolii',4:'yellowLeafCurlVirus'}

Path = "D:/ST/trainData/Tomato_pred"
imgPath = os.listdir(Path)

imgs=[]
pred_x = np.array([])
for fileName in imgPath:
    img = image.load_img(Path+'/'+fileName, target_size=(224, 224))
    imgs.append(img)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    pred_x = np.append(pred_x,x)
pred_x /=255.0
pred_x = pred_x.reshape(len(imgPath),1,224,224,3)
```

```python
print(len(imgPath))
```

```python
with tf.device("/device:CPU:0"):
    pxidx=0
    px = pred_x[pxidx]
    plt.figure()
    plt.imshow(imgs[pxidx])
    out = TomatoModel.predict(px)
    print("출력: ", out)
    maxm = 0
    maxmidx = 0
    for idx,value in enumerate(out[0]):
        if value > maxm:
            maxm = value
            maxmidx = idx
    print("예상 결과 : ",category[maxmidx])
```
